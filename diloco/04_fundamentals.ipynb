```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
T_seq = 10  # Sequential part execution time
T_par = 90  # Parallelizable part execution time
world_sizes = np.arange(1, 10001)  # Processor world sizes from 1 to 10000

# Compute execution time and speedup
T_P = T_seq + (T_par / world_sizes)  # Total execution time with P processors
S = (T_seq + T_par) / T_P  # Speedup

# Plot speedup vs. world size
plt.figure(figsize=(10, 6))
plt.plot(world_sizes, S, label="Speedup (S)", linewidth=2)
plt.title("Speedup vs. Number of Processors", fontsize=16)
plt.xlabel("Number of Processors (P)", fontsize=14)
plt.ylabel("Speedup (S)", fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
