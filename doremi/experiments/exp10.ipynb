```python
def print_name_weights(names, weights):
    for name, weight in zip(names, weights):
        print(name, weight)

    print(f"----- sorthing bellow ---")
    
    for idx in torch.argsort(weights, descending=True):
        print(names[idx], weights[idx], idx)
```

```python
def plot(names, weights):
    import matplotlib.pyplot as plt

    # Create bar chart
    plt.figure(figsize=(10, 8))
    plt.barh(names, weights.tolist(), color='skyblue')
    plt.xlabel('Weights')
    plt.ylabel('Names')
    plt.title('Bar Chart of Names vs Weights')
    plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
    
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot(names, doremi_weights, title):
    token_ratio_weights = [0.19948968279257367, 0.1, 0.1315255636357564, 0.05329604659793043, 0.05235711072309182,
                           0.047881985400873926, 0.046040384052077846, 0.04514776384899792, 0.044450375515503016,
                           0.0430242438158063, 0.042476733556139346, 0.0397504743690409, 0.039539991287963636,
                           0.039476812365103114, 0.037790322063391654, 0.037752509975750054]
    
    uniform_weights = [1/16] * 16
    
    # Normalizing Doremi weights for plotting
    doremi_weights_norm = [float(i) for i in doremi_weights]
    
    # Plot
    fig, ax = plt.subplots(figsize=(14, 8))
    
    barWidth = 0.25
    r1 = np.arange(len(names))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    
    ax.bar(r1, token_ratio_weights, color='b', width=barWidth, edgecolor='grey', label='Token Ratio Weights')
    ax.bar(r2, uniform_weights, color='r', width=barWidth, edgecolor='grey', label='Uniform Weights')
    ax.bar(r3, doremi_weights_norm, color='g', width=barWidth, edgecolor='grey', label='Doremi Weights')
    
    ax.set_xlabel('names', fontweight='bold')
    ax.set_xticks([r + barWidth for r in range(len(names))])
    ax.set_xticklabels(names, rotation=45, ha='right')
    ax.set_ylabel('Weights', fontweight='bold')
    ax.set_title(title)
    
    plt.legend()
    plt.tight_layout()
    plt.show()
    return plt
```

```python
plot(names, weights_100k.tolist(), title="DoReMi Weights at 100k")
```

```python
plot(names, weights_200k.tolist(), title="DoReMi Weights at 200k")
```
