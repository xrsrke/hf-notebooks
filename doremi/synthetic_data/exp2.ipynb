```python
import torch
```

```python
names = [
    "textbooks", "stories-ultrachat", "stories-openhermes", "jupyter", "amt-web", "amt-python", "amt-khan"
]
```

```python
weights_200k = torch.tensor([0.7032, 0.0620, 0.0566, 0.0443, 0.0650, 0.0276, 0.0413])
```

```python
for name, weight in zip(names, weights_200k):
    print(name, weight)
```

```python
reference_weights = torch.tensor([0.6269189655199215, 0.057163905092523316, 0.03981896616334453, 0.10559186614161543, 0.08580403968831365, 0.03644362961551115, 0.04825862777877037])
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt

# Setting the positions and width for the bars
positions = np.arange(len(names))
width = 0.35

# Plotting both the doremi_weights and reference_weights
plt.figure(figsize=(10, 8))
plt.bar(positions - width/2, weights_200k, width, label='Doremi Weights')
plt.bar(positions + width/2, reference_weights, width, label='Token ratio Weights')

plt.ylabel('Weights')
plt.title('Comparison of Doremi Weights and Reference Weights')
plt.xticks(positions, names, rotation='vertical')
plt.legend()

plt.tight_layout()
plt.show()
```
