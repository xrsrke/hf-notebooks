```python
import numpy as np
import matplotlib.pyplot as plt

def adam_simulation(iterations, learning_rate, beta1, beta2, epsilon, gradient):
    m = 0  # First moment estimate
    v = 0  # Second moment estimate
    theta = 0  # Parameter value
    
    history = []
    
    for t in range(1, iterations + 1):
        # Update biased first moment estimate
        m = beta1 * m + (1 - beta1) * gradient
        
        # Update biased second raw moment estimate
        v = beta2 * v + (1 - beta2) * (gradient ** 2)
        
        # Compute bias-corrected first moment estimate
        m_hat = m / (1 - beta1**t)
        
        # Compute bias-corrected second raw moment estimate
        v_hat = v / (1 - beta2**t)
        
        # Update parameters
        theta += learning_rate * m_hat / (np.sqrt(v_hat) + epsilon)
        
        history.append(theta)
    
    return history

# Set parameters
iterations = 70_000
learning_rate = 0.0006
beta1 = 0.9
beta2 = 0.95
epsilon = 1e-8
gradient = 0.00001

# Run simulation
results = adam_simulation(iterations, learning_rate, beta1, beta2, epsilon, gradient)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(range(1, iterations + 1), results)
plt.title('Adam Optimizer Simulation (Gradient = 0.001)')
plt.xlabel('Iterations')
plt.ylabel('Parameter Value')
plt.grid(True)
plt.show()

print(f"Final parameter value: {results[-1]:.6f}")
```
