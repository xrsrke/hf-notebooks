```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
TOKEN_PER_PARAMETER = 20  # Optimal tokens per parameter according to Chinchilla scaling
MAX_BATCH_SIZE = 16e6     # Maximum effective batch size in tokens

# Model sizes (number of parameters) from 1e9 to 1e12
N_parameters = np.logspace(9, 12, num=10)  # 10 points between 1e9 and 1e12

# Lists to store computed values
dataset_sizes = []
num_steps = []
batch_sizes = []

for N in N_parameters:
    # Optimal dataset size according to Chinchilla scaling
    D = TOKEN_PER_PARAMETER * N  # Total training tokens
    dataset_sizes.append(D)
    
    # Batch size (limited by MAX_BATCH_SIZE)
    batch_size = min(MAX_BATCH_SIZE, D)
    batch_sizes.append(batch_size)
    
    # Number of training steps
    steps = D / batch_size
    num_steps.append(steps)

# Convert lists to numpy arrays for plotting
N_parameters = np.array(N_parameters)
dataset_sizes = np.array(dataset_sizes)
num_steps = np.array(num_steps)

# Plotting
plt.figure(figsize=(12, 5))

# Plot Dataset Size vs. Model Size
plt.subplot(1, 2, 1)
plt.loglog(N_parameters, dataset_sizes, marker='o')
plt.title('Dataset Size vs. Model Size')
plt.xlabel('Model Size (Number of Parameters)')
plt.ylabel('Dataset Size (Number of Tokens)')
plt.grid(True, which="both", ls="--")

# Plot Number of Steps vs. Model Size
plt.subplot(1, 2, 2)
plt.loglog(N_parameters, num_steps, marker='o')
plt.title('Number of Training Steps vs. Model Size')
plt.xlabel('Model Size (Number of Parameters)')
plt.ylabel('Number of Training Steps')
plt.grid(True, which="both", ls="--")

plt.tight_layout()
plt.show()
```
